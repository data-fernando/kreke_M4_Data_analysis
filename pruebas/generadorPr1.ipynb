```python
import pandas as pd
import numpy as np
import random

# 1. Configuración de productos y sus bases lógicas
productos_base = {
    "Tablet": {"ID": 8, "Cat": "Electronica", "Precio": 300.0},
    "Cargador inalambrico": {"ID": 5, "Cat": "Electronica", "Precio": 40.0},
    "Smartphone": {"ID": 1, "Cat": "Electronica", "Precio": 600.0},
    "Reloj": {"ID": 7, "Cat": "Accesorios", "Precio": 150.0},
    "Zapatos deportivos": {"ID": 9, "Cat": "Calzado", "Precio": 85.0},
    "Pantalones": {"ID": 4, "Cat": "Ropa", "Precio": 45.0},
    "Bicicleta": {"ID": 10, "Cat": "Accesorios", "Precio": 400.0},
    "Chaqueta de cuero": {"ID": 2, "Cat": "Ropa", "Precio": 120.0},
    "Camiseta": {"ID": 6, "Cat": "Ropa", "Precio": 25.0},
    "Auriculares Bluetooth": {"ID": 3, "Cat": "Electronica", "Precio": 60.0}
}

n_filas = 500
data = []

# 2. Generación de datos
nombres_empleados = ["Pedro Gomez", "Jose Lopez", "Lucia Martinez", "Ana Garcia", "Carlos Perez"]
ciudades = ["Quito", "Guayaquil", "Cuenca", "Ambato", "Loja"]
metodos = ["Tarjeta de credito", "Efectivo", "PayPal", "Transferencia", "Tarjeta de debito"]

for i in range(n_filas):
    prod_nombre = random.choice(list(productos_base.keys()))
    info = productos_base[prod_nombre]
    
    precio_u = info["Precio"]
    cantidad = random.randint(1, 10)
    descuento_pct = random.choice([0, 5, 10, 15, 20])
    
    # CÁLCULO LÓGICO: Precio * Cantidad - Descuento
    monto_bruto = precio_u * cantidad
    monto_neto = monto_bruto * (1 - descuento_pct/100)
    
    data.append({
        "ID_venta": f"V{1000 + i}",
        "Sucursal": "Sucursal 1",
        "Fecha": pd.to_datetime("2023-01-01") + pd.to_timedelta(random.randint(0, 364), unit='d'),
        "ID_producto": info["ID"],
        "Nombre_producto": prod_nombre,
        "Categoría_producto": info["Cat"],
        "Precio_unitario": precio_u,
        "Cantidad_vendida": cantidad,
        "Monto_total": round(monto_neto, 2),
        "ID_empleado": f"E{random.randint(1, 5)}",
        "Nombre_empleado": random.choice(nombres_empleados),
        "Evaluación_atención": random.randint(1, 5),
        "Descuento": descuento_pct,
        "Ubicación_cliente": f"Zona {random.randint(1, 3)}",
        "Ciudad": random.choice(ciudades),
        "Método_pago": random.choice(metodos),
        "Hora_venta": random.choice(["Mañana", "Tarde", "Noche"])
    })

df_limpio = pd.DataFrame(data)

# Convertir tipos como tu estructura original
df_limpio['Categoría_producto'] = df_limpio['Categoría_producto'].astype('category')
df_limpio['Método_pago'] = df_limpio['Método_pago'].astype('category')

# 1. Ordenamos todo el DataFrame por fecha de forma ascendente
df_limpio = df_limpio.sort_values(by='Fecha').reset_index(drop=True)

# 2. (Opcional) Si quieres que los ID_venta también queden en orden con las fechas:
df_limpio['ID_venta'] = [f"V{1000 + i}" for i in range(len(df_limpio))]

# Guardar a CSV
df_limpio.to_csv("ventas_limpias.csv", index=False,encoding='utf-8-sig')
print("DataFrame generado con éxito.")
```
