```python
df['Fecha']=pd.to_datetime(df['Fecha'])
df['Categoría_producto']=pd.Categorical(df['Categoría_producto'])
df['Método_pago']=pd.Categorical(df['Método_pago'])

```

```python
## datos se ven muy inconsistentes
# Agrupamos por producto y categoría para ver la variación de precios
df_inconulta = df.groupby(["Nombre_producto", "Categoría_producto"])["Precio_unitario"].agg(["min", "max", "mean", "std"])

print(df_inconulta)

```

```python
# df_inconulta2 = df.groupby(["Nombre_producto", "ID_producto"])["Categoría_producto"].count()

# df_inconulta2

df_inconulta2 = df.groupby(["Nombre_producto", "ID_producto"]).agg(
    categoriaCount=("Categoría_producto","count"),
    precio_promedio=("Precio_unitario", "mean"),
    varianza_precio=("Precio_unitario", "var")
)

df_inconulta2
```
