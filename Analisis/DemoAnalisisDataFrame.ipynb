```python
import pandas as pd

data = {
    'Fecha': ['2021-01-01', '2021-01-02', '2021-01-03', '2021-01-04'],
    'Categoría del Producto': ['Electrónica', 'Ropa', 'Electrónica', 'Alimentos'],
    'Precio': [500, 80, 450, 20],
    'Cantidad': [1, 3, 2, 5],
    'Método de Pago': ['Tarjeta de Crédito', 'PayPal', 'Efectivo', 'Tarjeta de Crédito'],
    'Ubicación': ['Quito', 'Cuenca', 'Quito', 'Guayaquil']
}

df = pd.DataFrame(data)

print(df.info())

df["Categoría del Producto"]=pd.Categorical(df['Categoría del Producto'])
df["Método de Pago"]=pd.Categorical(df['Método de Pago'])


print(df.dtypes)
print(df.info())

```

```python
# extraer categorias en un dataframe, coloca columnas de bolleanos para cada categoria
df_categorias=pd.get_dummies(df,columns=["Categoría del Producto","Método de Pago"])
print("\n",df_categorias)
```

```python
# convertir a fechas
df["Fecha"]=pd.to_datetime(df["Fecha"])

df["UltimaCompraFecha"]=pd.to_datetime(['2021-02-02','2021-02-05','2021-02-04','2021-02-05'])

df["dias_desde_ultimaCompra"]=(df["UltimaCompraFecha"]-df["Fecha"]).dt.days

print(df.info())
```

## merge y concatenar

```python
data_promociones={
    'Categoría del Producto':['Electrónica', 'Ropa',  'Alimentos'],
    'Descuento':[10,5,2]
}

df_promociones=pd.DataFrame(data_promociones)

df_concatenado=pd.merge(df,df_promociones,on='Categoría del Producto',how="left")

print("talba combinada: \n",df_concatenado)
```
