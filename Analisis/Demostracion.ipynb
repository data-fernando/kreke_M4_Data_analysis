## Fechas

```python
import pandas as pd
```

```python

ventas = {
    "ventas": [100, 120, 90, 150, 130],
    "producto": ["A", "B", "A", "C", "B"]
}

# Creamos una lista de 5 fechas (una por cada registro)
fechas = pd.to_datetime([
    '2025-04-02', '2025-04-03', '2025-04-04', '2025-04-05', '2025-04-06'
])

# Convertimos a DataFrame
df = pd.DataFrame(ventas)

# Añadimos la columna de fechas
df["fecha"] = fechas

print(df)
```

```python
import random
df["fecha_creacion"] = pd.to_datetime(["2024-04-02"] * len(df))
df["aleatorio"]=[random.choice([True, False]) for _ in range(len(df))]
print(df)
```

.loc[] → selecciona filas por etiquetas (nombres de índice o condiciones).
Ejemplo:

```python
df.loc[3]              # fila con índice 3
df.loc[df["columna"] > 20]  # todas las filas donde columna > 20


```

.iloc[] → selecciona filas y columnas por posición numérica (como si fueran coordenadas).
Ejemplo:

```python
df.iloc[0]             # primera fila
df.iloc[:, 1]          # toda la segunda columna
df.iloc[2, 1]          # valor en fila 3, columna 2
```
