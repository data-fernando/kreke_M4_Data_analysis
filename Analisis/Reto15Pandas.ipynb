```python


data3 = {
    "Producto": ["Laptop", "Celular", "Tablet", "Auriculares", "Teclado"],
    "Precio": [500, 300, 150, 50, 75],
    "Cantidad Vendida": [10, 15, 7, 25, 30],
    # 2 convertir fecha
    "Fecha de Venta": pd.to_datetime(
        ["2025-04-01", "2025-04-03", "2025-04-05", "2025-04-06", "2025-04-10"]
    )
}


df_ventas = pd.DataFrame(data3)

print(df_ventas)

# 3 
df_ventas["ingreso_total"]=df_ventas["Precio"]*df_ventas["Cantidad Vendida"]
print("\nagregar ingreso total: \n",df_ventas)

# 4 filtro mayor a 100
df_ventas_1000=df_ventas[df_ventas["ingreso_total"]>1000]
print("\nmayor a 1000: \n",df_ventas_1000)


# 5 mes de fecha de venta
df_ventas["mes"]=df_ventas["Fecha de Venta"].dt.month
print("\n agregar mes de venta: \n",df_ventas)

# 6 filtro abril
df_ventas_filtro_abril=df_ventas[df_ventas["mes"]==4]
print("\n filtro abril: \n",df_ventas)
```

## ejercicio 4


```python

from datetime import date

hoy = date.today()
# "2026-04-13"

fechas = pd.date_range(start=hoy, periods=7, freq="D")
temperaturasCel = [18.5, 20.3, 22.0, 21.8, 19.7, 20.1, 21.4]

df_ejercicio4 = pd.DataFrame({
    "Fecha": fechas,
    "TemperaturaCelcius": temperaturasCel
})

print(df_ejercicio4)

# 2 
df_filtrado = df_ejercicio4[df_ejercicio4["Fecha"].dt.weekday.isin([0, 2, 4])]
print("\n2. lunes , miercoles , viernes filtrado \n",df_filtrado["TemperaturaCelcius"])

# 3
print("\n3. media demanal \n",float(df_ejercicio4["TemperaturaCelcius"].mean()))

# 4 y 5
df_ejercicio4["TemperaturaFahrenheit"] = df_ejercicio4["TemperaturaCelcius"] * 1.8 + 32

print("\n4 y 5.DF con series ferenheit: \n",df_ejercicio4)

# 6 
# Obtener fila con temperatura maxima
fila_max = df_ejercicio4.loc[df_ejercicio4["TemperaturaFahrenheit"].idxmax()]
diatempAlta = fila_max["Fecha"]

# Obtener fila con temperatura minima
fila_min = df_ejercicio4.loc[df_ejercicio4["TemperaturaFahrenheit"].idxmin()]
diatempBaja = fila_min["Fecha"]
print("6 fechas de dias maximos")
print("Día con mayor temperatura:", diatempAlta)
print("Día con menor temperatura:", diatempBaja)

```

## ejercicio 5


```python

df_estudiantes = pd.DataFrame({
    "Estudiante": ["Ana", "Luis", "Carlos", "Maria", "Pedro"],
    "Días Asistidos": [20, 18, 22, 15, 19],
    "Mes": ["Abril", "Abril", "Abril", "Abril", "Abril"]
})

print(df_estudiantes)

df_estudiantes["dias_mes"]=df_estudiantes["Mes"].apply(lambda x:22 if x=="Abril" else 23 )

print("\n2. agregar dias de mes\n",df_estudiantes)

df_estudiantes["asistencia_porcentaje"]=(df_estudiantes["Días Asistidos"]/df_estudiantes["dias_mes"])*100
df_estudiantes_filtro80Asistencia=df_estudiantes[df_estudiantes["asistencia_porcentaje"]>=80]
print("\n4 y 5. filtrar asistencis >80%\n",df_estudiantes_filtro80Asistencia)


print("\n5. ordenar decendentemente\n",df_estudiantes["asistencia_porcentaje"].sort_values(ascending=False))
```

```python
df_finanzas = pd.DataFrame({
    "Fecha": pd.to_datetime(["2025-03-01", "2025-03-05", "2025-03-10", "2025-03-15", "2025-03-20"]),
    "Ingreso": [1500, 1200, 2000, 1800, 1600],
    "Egreso": [500, 700, 600, 800, 400]
})

print(df_finanzas)

# 3


df_finanzas["saldo_diario"]=df_finanzas["Ingreso"]-df_finanzas["Egreso"]
print("\n3. saldo:\n",df_finanzas)

# Saldo acumulado
df_finanzas["saldo_acumulado"] = df_finanzas["saldo_diario"].cumsum()

print("\n4. saldo acumulado:\n", df_finanzas)

# 5
#logica personalizada
df_finanzas_filtro=df_finanzas.loc[df_finanzas["saldo_diario"]>0 ]

# mostrar mas de una fila
print("\n5. mostras fechas y saldo:\n", df_finanzas[["Fecha","saldo_diario"]])

```
