## Funcion Apply

````python
import pandas as pd

df=pd.DataFrame({
    "nombres":["marko","pepe","gonzalo"],
    "edades":[19,18,16]
})

df_demo=df.copy()

def multiplicarx2(valorinicial):
    return valorinicial*2

df_demo["multiplicadoX2"]=df_demo["edades"].apply(multiplicarx2)

print(df_demo)

````

## editar un registro y agrupar un set

````python
## basandose en el ejercicio anterios

df_demo["ciudad"]=df_demo["nombres"].map({"marko":"quito","pepe":"guayaquil","gonzalo":"rosario"})

print(df_demo)

grupo_edades=df_demo.groupby("edades")


# contrar
print(f"\n {type(grupo_edades)}\n",grupo_edades.size())
````

## Limpiar ,rellenar Valores



#### usar el dataframe anterior para avanzar con los cambios 

````python
## quitar valores nulos 



# Crear DataFrame con valores nulos y datos "sucios"
import pandas as pd
import numpy as np

df_sucio = pd.DataFrame({
    "producto": ["Laptop", "Celular", "Celular", "Celular", "Tablet", "Laptop",
                 "Monitor", "Mouse", "Impresora", None, None, "Impresora"],
    "precio": [1200, None, None, 800, 500, 80,
               300, 40, 50, None, None, 200],
    "stock": [10, 8, 8, "n/a", 5, 2,
              40, 15, 20, "N/A", "N/A", 3],
    "categoria": ["Electrónica", "Electrónica", "Electrónica", "Electrónica", "Electrónica", "Electrónica",
                  "Electrónica", None, "Accesorios", "Accesorios", "Accesorios", None]
})

print("=== Data original ===")
print(df_sucio)

# df_limpio = df_sucio.dropna(how="all")

# remplazar la columna los valores raros o problematicos con valores conocidoa
# usar el dataframe anterior para avanzar con los cambios 
df_rellenado=df_sucio.copy()
df_rellenado["stock"] = df_rellenado["stock"].replace(["N/A", "n/a"], np.nan)
df_rellenado=df_rellenado.fillna({"stock":0})
df_rellenado=df_rellenado.fillna({"producto":"Desconocido"})
df_rellenado=df_rellenado.fillna({"categoria":"Otros"})

print("\nvalores rellenos:\n",df_rellenado)

````

````python

df_sinVacios=df_sucio.dropna()
print("\nCorregido sin nulos:\n",df_sinVacios)

df_sinVacios["stock"] = df_sinVacios["stock"].replace(["n/a", "N/A"], 0)

print("\nCorregido reemplazados:\n",df_sinVacios)
````

````python
df_sinduplicados=df_sucio.drop_duplicates()

print(f"\ncon duplicados: {len(df_sucio)}\n",df_sucio, )
print(f"\nSin duplicados: {len(df_sinduplicados)}\n",df_sinduplicados)

````
